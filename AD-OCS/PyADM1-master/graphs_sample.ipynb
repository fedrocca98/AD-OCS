```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
name = 'bsm2'
influent_state = pd.read_csv(name  + '_digester_influent.csv')
effluent_state = pd.read_csv(name +  "_dynamic_out.csv")
gas = pd.read_csv(name + "_gas_flows.csv")
```

```python
tspan = influent_state['time']
```

```python


plt.figure(2)
plt.plot(tspan, effluent_state['S_su'], label='S_su')
plt.plot(tspan, effluent_state['X_su'], label='X_su')
plt.legend(loc='best')
plt.figure(3)
plt.plot(tspan, effluent_state['S_aa'], label='S_aa')
plt.plot(tspan, effluent_state['X_aa'], label='X_aa')
plt.legend(loc='best')
plt.figure(4)
plt.plot(tspan, effluent_state['S_fa'], label='S_fa')
plt.plot(tspan, effluent_state['X_fa'], label='X_fa')
plt.legend(loc='best')

plt.figure(5)
plt.subplot(2,1,1)
plt.plot(tspan, gas['q_ch4'], label='CH4')

plt.legend(loc='best')
plt.subplot(2,1,2)
plt.plot(tspan, gas['q_gas'], label='CO2')
plt.legend(loc='best')
#plt.show()
```
